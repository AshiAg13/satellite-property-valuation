```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("seaborn-v0_8")
```

```python
# Load raw data
train = pd.read_csv("data/raw/train.csv")
test = pd.read_csv("data/raw/test.csv")

print(train.shape)
print(test.shape)

train.head()
```

```python
# Basic info check
train.info()
train.describe()
```

```python
# Check column names
print(train.columns)


# Remove duplicate
train = train.drop_duplicates()
test = test.drop_duplicates()

print("Train shape after dedup:", train.shape)
print("Test shape after dedup:", test.shape)

```

```python
# Columns to drop
drop_cols = ['id', 'date', 'zipcode', 'yr_renovated', 'floors']

train_clean = train.drop(columns=drop_cols)
test_clean = test.drop(columns=drop_cols)

print(train_clean.shape)
print(test_clean.shape)
```

```python

# Feature Engineering
# The goal here is to create more informative and interpretable features
# from existing columns, without overcomplicating the model.

CURRENT_YEAR = 2015  # Approximate year of dataset collection

# 1. House age (more meaningful than raw year built)
train_clean['house_age'] = CURRENT_YEAR - train['yr_built']
test_clean['house_age'] = CURRENT_YEAR - test['yr_built']

# 2. Basement ratio
# Captures how much of the living space is underground
train_clean['basement_ratio'] = train['sqft_basement'] / (train['sqft_living'] + 1)
test_clean['basement_ratio'] = test['sqft_basement'] / (test['sqft_living'] + 1)

# 3. Living area compared to neighborhood average
# Positive value means the house is larger than nearby houses
train_clean['living_vs_neighbors'] = train['sqft_living'] - train['sqft_living15']
test_clean['living_vs_neighbors'] = test['sqft_living'] - test['sqft_living15']

# 4. Luxury indicator based on construction grade
# Higher-grade houses generally command a price premium
train_clean['luxury_grade'] = (train['grade'] >= 10).astype(int)
test_clean['luxury_grade'] = (test['grade'] >= 10).astype(int)

# Check final feature shape
print("Train shape after feature engineering:", train_clean.shape)
print("Test shape after feature engineering:", test_clean.shape)
```

```python
plt.figure(figsize=(6,4))
plt.hist(train_clean['price'], bins=50)
plt.xlabel("Price")
plt.ylabel("Count")
plt.title("House Price Distribution")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(train_clean['sqft_living'], train_clean['price'], alpha=0.3)
plt.xlabel("Sqft Living")
plt.ylabel("Price")
plt.title("Living Area vs Price")
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(train_clean.corr(), cmap='coolwarm', linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
train_clean.to_csv("data/processed/train_clean.csv", index=False)
test_clean.to_csv("data/processed/test_clean.csv", index=False)
```
